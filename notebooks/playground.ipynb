```python
import re
from pathlib import Path
from collections import defaultdict
```

```python
SUBJECT_FOLDER_PATTERN = "S_%d"
RE_SUBJECT_FOLDER_PATTERN = "S_\\d+"
GLOB_SUBJECT_FOLDER_PATTERN = "S_*"

TRIAL_FOLDER_PATTERN = "Trial_%d"
RE_TRIAL_FOLDER_PATTERN = "Trial_\\d+"
GLOB_TRIAL_FOLDER_PATTERN = "Trial_*"

DATASET_PATH = Path("../data/Data_Pattern/")

exclude_samples = {"S_1": [], "S_2": ["Trial_1"]}
```

```python
paths = []
for path in DATASET_PATH.glob(f"{GLOB_SUBJECT_FOLDER_PATTERN}/{GLOB_TRIAL_FOLDER_PATTERN}"):
    for subject, trials in exclude_samples.items():
        if re.search(f"{subject}/", path.as_posix()):
            if len(trials) == 0:
                continue

            for trial in trials:
                if re.search(f"{trial}$", path.as_posix()):
                    continue
    paths.append(path)
```

```python
def extract_keys(path: Path):
    path = path.as_posix()

    s = re.findall(RE_SUBJECT_FOLDER_PATTERN, path)
    t = re.findall(RE_TRIAL_FOLDER_PATTERN, path)

    if (not s) or (not t):
        raise ValueError("Could not be parsed %s" % path.as_posix())

    s = int(s[0].split("_")[1])
    t = int(t[0].split("_")[1])

    return s, t


sorted(paths, key=extract_keys)[:10]
```

```python
dataset_map = defaultdict[int, defaultdict[int, Path]](defaultdict)
for path in paths:
    s, t = extract_keys(path)
    dataset_map[s][t] = path
dataset_map
```
